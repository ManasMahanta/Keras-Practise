```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_csv=pd.read_csv("Google_Stock_Price_Train.csv")
```

```python
train_csv
```

```python
training_set=train_csv.iloc[:,1:2].values
training_set
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
scaled_training=scaler.fit_transform(training_set)
scaled_training
```

```python
X_train=[]
Y_train=[]

for i in range(60,1258):
    X_train.append(scaled_training[i-60:i,0])
    Y_train.append(scaled_training[i,0])
X_train,Y_train=np.array(X_train),np.array(Y_train)    
```

```python
X_train
```

```python
X_train=np.reshape(X_train,(X_train.shape[0],X_train.shape[1],1)) #reshaping for inputs
```

```python
from keras.layers import Dense
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dropout

import warnings
warnings.filterwarnings('ignore')
```

```python
model=Sequential()
model.add(LSTM(units=50,return_sequences=True,input_shape=(X_train.shape[1],1)))
```

```python
model.add(Dropout(0.2))
```

```python
model.add(LSTM(units=50,return_sequences=True))
```

```python
model.add(Dropout(0.2))
```

```python
model.add(LSTM(units=50,return_sequences=True))
```

```python
model.add(Dropout(0.2))
```

```python
model.add(LSTM(units=50))
```

```python
model.add(Dropout(0.2))
```

```python
model.add(Dense(units=1))
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(X_train,Y_train,batch_size=32,epochs=100)
```

```python
test_csv=pd.read_csv("Google_Stock_Price_Test.csv")
```

```python
dataset_total=pd.concat((train_csv['Open'],test_csv['Open']),axis=0)
```

```python
input_dataset=dataset_total[len(dataset_total)-len(test_csv)-60:].values
```

```python
input_dataset=input_dataset.reshape(-1,1)
```

```python
scaled_test=scaler.transform(input_dataset)
```

```python
X_test=[]
for i in range(60, 80):
    X_test.append(scaled_test[i-60:i, 0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
predicted_stock_price = model.predict(X_test)
```

```python
predicted_stock_price = scaler.inverse_transform(predicted_stock_price)
```

```python
real_stock_price = test_csv.iloc[:, 1:2].values
```

```python
plt.plot(real_stock_price, color = 'red', label = 'Real Google Stock Price')
plt.plot(predicted_stock_price, color = 'blue', label = 'Predicted Google Stock Price')
plt.title('Google Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Google Stock Price')
plt.legend()
plt.show()
```
